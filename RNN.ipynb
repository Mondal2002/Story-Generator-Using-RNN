```python
import os

def get_all_txt_files(folder_path):
  """Returns a list of all the text files in the given folder path."""
  txt_files = []
  for root, dirs, files in os.walk(folder_path):
    for file in files:
      if file.endswith(".txt"):
        txt_files.append(os.path.join(root, file))
  return txt_files

# Get the path to the folder containing the text files
folder_path = "C:/Users/Subham Mondal/Desktop/sherlockdataset/sherlock/sherlock"

# Get a list of all the text files in the folder
txt_files = get_all_txt_files(folder_path)

# Print the list of text files
for txt_file in txt_files:
  print(txt_file)
```

```python
import os

def copy_all_txt_files_to_one(folder_path, output_file_path):
  """Copies all the txt files in the given folder to the given output file."""
  with open(output_file_path, "w") as output_file:
    for txt_file in get_all_txt_files(folder_path):
      with open(txt_file, "r") as input_file:
        output_file.write(input_file.read())

# Get the path to the folder containing the text files
folder_path = "C:/Users/Subham Mondal/Desktop/sherlockdataset/sherlock/sherlock"

# Get the path to the output file
output_file_path = "C:/Users/Subham Mondal/Desktop/Subham Mondal/sherlockcompile.txt"

# Copy all the txt files to the output file
copy_all_txt_files_to_one(folder_path, output_file_path)

```
